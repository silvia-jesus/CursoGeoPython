Álgebra de mapas refere-se a um conjunto de operações que podem ser aplicadas a dados raster para gerar novos dados. Por exemplo, somar duas imagens raster ou calcular índices ambientais como o NDVI.

* Operações matemáticas: Soma, subtração, multiplicação, divisão entre pixels de um ou mais rasters.
* Operações lógicas: Comparações entre valores de pixels para identificar regiões que atendam a certas condições.

```python
!pip install rasterio
import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt
```

Carregando o arquivo empilhado

```python
path ='/content/drive/MyDrive/DADOS_GEOPYTHON/Aula05/SAIDA/L8_2020075_20240821_B1234567.tif'
```

```python
# Abrir o arquivo empilhado
with rio.open(path) as src:
    # Ler as bandas (4 - Vermelho, 3 - Verde, 2 - Azul)
    nir = src.read(5)  # Banda 5 (NIR)
    red = src.read(4)   # Banda 4 (Vermelho)
```

Exemplos de Álgebra de Mapas

*Índice de Vegetação da Diferença Normalizada (NDVI)*

Um exemplo comum de álgebra de mapas é o cálculo do NDVI, que é uma medida da saúde da vegetação. O NDVI é calculado da seguinte forma:

image.png

image.png

```python
# Calcular NDVI
ndvi = (nir.astype(np.float32) - red.astype(np.float32)) / (nir.astype(np.float32) + red.astype(np.float32))
```

```python
# Exibir o NDVI
plt.figure(figsize=(10, 10))
# plt.imshow(ndvi, cmap='RdYlGn', vmin=-1, vmax=1)  # Usando o intervalo original de -1 a 1
plt.imshow(ndvi, cmap='RdYlGn')
plt.title('NDVI - Índice de Vegetação da Diferença Normalizada')
plt.axis('off')
plt.colorbar(label='NDVI')
plt.show()
```
